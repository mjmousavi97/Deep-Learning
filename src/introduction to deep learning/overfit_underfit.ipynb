```python
print(f"Number of all sample in dataset: {x.shape[0]}")
print(f"Number of features: {x.shape[1]}")
print("-"*40)
print(f"Shape of x train: {x_train.shape}")
print(f"Shape of x test: {x_test.shape}")

print(f"Shape of y train: {y_train.shape}")
print(f"Shape of y test: {y_test.shape}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=3)

knn_clf.fit(x_train, y_train)
y_pred = knn_clf.predict(x_test)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))
neighbors = [1, 3, 9]

for neighbor, ax in zip(neighbors, axes):
    clf = KNeighborsClassifier(n_neighbors=neighbor).fit(x, y)
    mglearn.plots.plot_2d_separator(classifier=clf, X=x, fill=True, eps=0.5, ax=ax, alpha=0.4)
    mglearn.discrete_scatter(x[:, 0], x[:, 1], y, ax=ax)
    ax.set_title(f"neighbor: {neighbor}")
```

# For the second dataset(breast canser)

```python
from sklearn.model_selection import train_test_split

x, y = (breast_data['data'], breast_data['target'])
x_train, x_test, y_train, y_test = train_test_split(x, y)

print(f"Number of all sample in dataset: {x.shape[0]}")
print(f"Number of features: {x.shape[1]}")
print("-"*40)
print(f"Shape of x train: {x_train.shape}")
print(f"Shape of x test: {x_test.shape}")

print(f"Shape of y train: {y_train.shape}")
print(f"Shape of y test: {y_test.shape}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = np.arange(10) + 1
test_scores = []
train_scores = []


for neighbor in neighbors:
    clf = KNeighborsClassifier(n_neighbors=neighbor).fit(x_train, y_train)
    test_score = clf.score(x_test, y_test)
    train_score = clf.score(x_train, y_train)
    test_scores.append(test_score)
    train_scores.append(train_score)

plt.plot(neighbors, test_scores)
plt.plot(neighbors, train_scores)
```
