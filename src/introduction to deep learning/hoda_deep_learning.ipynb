```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Input
import numpy as np
import matplotlib.pyplot as plt

from load_dataset_hoda import load_hoda
```

```python
x_train, y_train, x_test, y_test = load_hoda(data_path="", train_size=1000, test_size=200, img_size=(5, 5))

print(f"Shape of x train: {x_train.shape}")
print(f"Shape of x test: {x_test.shape}")

print(f"Shape of y train: {y_train.shape}")
print(f"Shape of y test: {y_test.shape}")
```

```python
y_train_cat = keras.utils.to_categorical(y_train, num_classes = 10)
y_test_cat = keras.utils.to_categorical(y_test, num_classes = 10)
```

```python
x_train = x_train.astype(dtype=np.float32)
x_test = x_test.astype(dtype=np.float32)

x_train /= 255
x_test /= 255
```

```python
model = Sequential()
model.add(Input(shape=(25, )))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
            optimizer='rmsprop',
            metrics=['accuracy']
)

# prefer to use : loss = tf.keras.losses.categorical_crossentropy(from_logits=True). remember to remove activation='softmax' from the last layer
# prefer to use : loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True). remember not to change y_train to one hot.
```

```python
history = model.fit(x_train, y_train,
                    epochs=100,
                    batch_size=64,
                    validation_split=0.2
)
```

```python
plt.plot(history.history['val_accuracy'])
plt.plot(history.history['accuracy'])

plt.legend(['Val, Train'])
```

```python
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])

plt.legend(['Val, Train'])
```

```python
loss, acc = model.evaluate(x_test, y_test)

print(f"Accuracy is {acc} and loss is {loss}.")
```
