# Split the dataset to train, test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris_dataset["data"], iris_dataset["target"], random_state=42)
```

```python
print(f"Shape of train set = {x_train.shape}")
print(f"Shape of test set = {x_test.shape}")
```

```python
df = pd.DataFrame(data=x_train, columns=iris_dataset["feature_names"])
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X=x_train, y=y_train)
```

```python
x_new_sample = np.array([[4, 6, 8, 0.3]])
x_new_sample.shape
```

```python
prediction_sample = knn.predict(X=x_new_sample)
print(prediction_sample)
print(f"Prediction of x sample is: {iris_dataset['target_names'][prediction_sample]}")
```

```python
y_pred = knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
y_pred == y_test
```

```python
print(f"Accuracy: {np.round_(np.mean(y_test == y_pred), 2) * 100}")
```

```python
print(f"Accuracy: {knn.score(X=x_test, y=y_test) * 100}")
```
