```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras

from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input, decode_predictions

from keras.preprocessing import image
# import keras.utils as image
```

```python
model = VGG16(weights="imagenet")
model.summary()
```

```python
image_path = "/mnt/c/Users/Mousavi/Downloads/deeplearning-tensorflow2-notebooks-master/deeplearning-tensorflow2-notebooks-master/image/elephant.jpg" 
image = image.load_img(image_path)
image_array = np.array(image)
print(f"Shape of image is: {image_array.shape}")

x = np.expand_dims(image_array, axis=0)
print(f"Shape of input dataset: {x.shape}")
```

```python
plt.imshow(image_array)
```

```python
x = preprocess_input(x)
preds = model.predict(x)
```

```python
max_confidence = np.max(preds)
class_num = np.argmax(preds)
```

```python
decode_predictions(preds)[0]
```
