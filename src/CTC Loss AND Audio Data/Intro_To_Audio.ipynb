```python
amp = np.iinfo(np.uint16).max
```

```python
x = np.linspace(0., 1., 16000)
y = amp / 2 * np.sin(2 * np.pi *  50 * x)
plt.xlim(0, 1)
plt.plot(x, y)
```

```python
IPython.display.Audio(y, rate=16000)
```

```python
x = np.linspace(0, 1, 16000)
y1 = amp / 2 * np.sin(2 * np.pi * 50 * x)
y2 = amp / 2 * np.sin(2 * np.pi * 200 * x)

y3 = y1 + y2
y4 = y3 / 2

plt.figure(figsize=(10, 20))
plt.subplot(4, 1, 1)
plt.plot(y1[:1000])


plt.subplot(4, 1, 2)
plt.plot(y2[:1000])

plt.subplot(4, 1, 3)
plt.plot(y3[:1000])

plt.subplot(4, 1, 4)
plt.plot(y4[:1000])
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 16000

# FFT
Y = np.fft.fft(y4)

freq = np.fft.fftfreq(len(y4), d=1/fs)

magnitude = np.abs(Y)
```

```python
plt.figure(figsize=(10,4))

plt.plot(freq[:len(freq)//2], magnitude[:len(magnitude)//2])

plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT of y4")
plt.grid()
plt.show()
```
