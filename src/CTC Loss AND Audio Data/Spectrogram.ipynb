```python
import scipy.io as sio
import IPython
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
```

```python
sample_rate, data = sio.wavfile.read("/mnt/e/Deep Learning/data/Recording.wav")
print(f"Sample rate is: {sample_rate}")
print(f"Number of Chanelss: {data.shape[1]}")
print(f"Duration of Audio: {data.shape[0] / sample_rate}'s")
```

```python
IPython.display.Audio(data[:, 0], rate=sample_rate)
```

```python
plt.plot(data[:, 1])
```

```python
mono = data.mean(axis=1)

plt.figure(figsize=(12, 5))

plt.specgram(
    mono,
    Fs=sample_rate,
    NFFT=1024,
    noverlap=512,
    cmap="viridis"
)

plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.title("Spectrogram")
plt.colorbar(label="Intensity")
plt.show()
```
